```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
t=np.linspace(-2*math.pi ,  2*math.pi, num=1000)
y=np.zeros(1000)
print(t.shape)
print(y.shape)
```

# approximate   f(x)=x^2/2

```python
for n in range(1,20):
    an=2*math.pow(-1,n)/n/n
    y=y +an*np.cos(n*t)
y=y+(math.pi)*(math.pi)/6
plt.plot(t,y)
plt.grid(True)
plt.show()
plt.savefig('fourser2.eps',format='eps',dpi=1000)
```
